```python
import requests
import bs4
import csv
```

```python
r=requests.get("https://open.163.com/ocw/#media")
```

```python
r.text
```

```python
from bs4 import BeautifulSoup
```

```python
mypage=BeautifulSoup(r.text)
```

```python
mypage
```

```python
cate=mypage.find_all("div",attrs={"class":"u-tag3"})
```

```python
cate
```

```python
categories=[]

for i in range (0,len(cate)):
    name=cate[i].find("h2")
    sub=cate[i].find("span")
    categories.append(name.text + " " + sub.text )
```

```python
categories
```

```python
vedio=mypage.find_all("div",attrs={"class":"m-cate m-cate1"})
```

```python
vedio
```

```python
titles=[]

for genre in vedio:
    title=genre.find_all('h5')
    headline=[]
    for i in title:
        headline.append(i.text)
    titles.append(headline)
```

```python
titles
```

```python
with open('class_data.csv','w') as f:
    writer = csv.writer(f)
    header = ['categories','vedios']
    writer.writerow(header)
    for i in range(0,len(categories)):
        
        a=[categories[i]] * len(titles[i]),titles[i]
        b=map(list,zip(*a))
        writer.writerows(b)
```
